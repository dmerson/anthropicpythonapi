```python
# Create an API client
from anthropic import Anthropic
client = Anthropic()
model = "claude-sonnet-4-6"
```

```python
def add_user_message(messages, text):
    messages.append({"role": "user", "content": text})

def add_assistant_message(messages, text):
    messages.append({"role": "assistant", "content": text})

def chat(messages):
    message = client.messages.create(
        model=model,
        max_tokens=1000,
        messages=messages,
    )
    return message.content[0].text

```

```python
#Make a starting list of messsages
messages = []

#Add in the initial user question of "Define quantum computing in one sentence
add_user_message(messages, "What is quantum computing? Answer in one sentence.")    

messages
#Pass the list of messages into chat to get a response from the model
answer=chat(messages)
#Take the answer and add it as an assistant message to the list of messages
add_assistant_message(messages, answer)
#Add in the user's follow up question "What are the applications of quantum computing?" as a user message to the list of messages
add_user_message(messages, "Write another sentence")
#Call chat again with the list of messages to get a response from the model
final_answer = chat(messages)
```
